```python
# import libreries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load the Datasets : UserDetails, CookingSessions, OrderDetails

```python
user_details = pd.read_excel("UserDetails.csv.xlsx")
user_details
```

```python
cooking_sessions = pd.read_excel("CookingSessions.csv.xlsx")
cooking_sessions
```

```python
order_details = pd.read_excel("OrderDetails.csv.xlsx")
order_details
```

```python
# Total information of the datasets
print(f"\nUser Details:\n {user_details.info()}")
print(f"\nCooking Sessions:\n {cooking_sessions.info()}")
print(f"\nOrder Details:\n {order_details.info()}")
```

#### Check the Datasets for missing values , Duplicate values

```python
# Find Missing Values
print(f"\nMissing Values in User Details: \n{user_details.isnull().sum()}")
print(f"\nMissing Values in Cooking Sessions:\n{cooking_sessions.isnull().sum()}")
print(f"\nMissing Values in Order Details:\n{order_details.isnull().sum()}")
```

There are no missing values available in Userdetails and Cooking session dataset. But there is 2 missing values available in order details dataset.

```python
# There are no duplicate entries
print(f"\nDuplicates in User Details: {user_details.duplicated().sum()}")
print(f"Duplicates in Cooking Sessions: {cooking_sessions.duplicated().sum()}")
print(f"Duplicates in Order Details: {order_details.duplicated().sum()}")
```

There are no duplicated values present.

#### Data Cleaning

```python
# Convert date fields to datetime format for consistency
user_details['Registration Date'] = pd.to_datetime(user_details['Registration Date'])
cooking_sessions['Session Start'] = pd.to_datetime(cooking_sessions['Session Start'])
cooking_sessions['Session End'] = pd.to_datetime(cooking_sessions['Session End'])
order_details['Order Date'] = pd.to_datetime(order_details['Order Date'])
```

```python
# Check columns available in datasets
print("OrderDetails Columns:", order_details.columns)
print("CookingSessions Columns:", cooking_sessions.columns)
print("UserDetails Columns:", user_details.columns)
```

#### Data Merging

```python
# Merge datasets: OrderDetails -> CookingSessions -> UserDetails
merged_data = order_details.merge(cooking_sessions, on="Session ID", how="left")
merged_data.head()
```

```python
# Keep User ID_x and drop User ID_y
intermediate_merged = merged_data.rename(columns={"User ID_x": "User ID"}).drop(columns=["User ID_y"])

print(intermediate_merged.columns)
```

Here, we need to column name consistent. So we will keep the USER ID_X as it has proper ID and drop USER ID_Y as it's all data will available in USER ID_X

```python
# Merge with UserDetails using the cleaned User ID
merged_data = intermediate_merged.merge(user_details, on="User ID", how="left")
merged_data.columns
```

```python
merged_data = merged_data.rename(columns={"Dish Name_x": "Dish Name"}).drop(columns=["Dish Name_y"])
merged_data = merged_data.rename(columns={"Meal Type_x": "Meal Type"}).drop(columns=["Meal Type_y"])

print(merged_data.columns)
```

```python
# Display the final merged data
merged_data.head()
```

#### Analysis

#### 1. Relationship between cooking sessions and orders

```python
## Check how many cooking sessions resulted in completed orders
completed_sessions = merged_data[merged_data['Order Status'] == "Completed"]
completed_ratio = len(completed_sessions) / len(cooking_sessions) * 100
print(f"\nPercentage of Cooking Sessions that Resulted in Completed Orders: {completed_ratio:.2f}%")
```

The 87.50% of cooking sessions that result in completed orders is a strong positive metric, indicating that most users who engage with cooking sessions find them valuable enough to follow through with an order. However, there’s still room for improvement, particularly in understanding and addressing the 12.5% of sessions that don’t lead to purchases. This analysis provides a foundation for improving both the cooking experience and overall order completion rate.

```python
## Correlation between session rating and order rating
rating_correlation = merged_data[['Session Rating', 'Rating']].corr()
print("\nCorrelation between Session Rating and Order Rating:")
print(rating_correlation)
```

The moderate positive correlation (0.635) between Session Rating and Order Rating implies that users who enjoy their cooking sessions tend to enjoy their orders more, suggesting that businesses could focus on improving session quality to boost overall user satisfaction and ratings.

#### 2. Popular Dishes

```python
## Top 5 most ordered dishes
top_dishes = completed_sessions['Dish Name'].value_counts().head(5)
print("\nTop 5 Dishes by Number of Orders:")
print(top_dishes)
```

The top 5 dishes by number of orders provide a clear picture of user preferences, with Spaghetti being the most popular dish, followed by Caesar Salad and Grilled Chicken. The data highlights both opportunities for promoting popular dishes and potential for growth in less-ordered categories like Oatmeal and Pancakes. Based on this analysis, we can strategize to improve meal offerings, target specific meal times, and cater to a broader range of customer preferences.

#### 3. Demographic Trends

```python
## Average orders per age group
age_orders = merged_data.groupby('Age')['Total Orders'].mean()
print("\nAverage Orders by Age Group:")
print(age_orders)
```

The analysis of average orders by age group reveals that users in the 35-42 age group are the highest orderers, while younger individuals (especially in the 25-27 range) tend to place fewer orders. This provides an opportunity to target promotions, loyalty programs, or meal plans toward older age groups, while also focusing on engaging younger users to increase their ordering frequency.

```python
## Favorite Meal Type by location
location_meal_preference = merged_data.groupby('Location')['Favorite Meal'].value_counts()
print("\nFavorite Meal Type by Location:")
print(location_meal_preference)
```

This analysis reveals clear regional trends in meal preferences, with Dinner being the most popular meal across several cities, while Lunch and Breakfast are also favored in specific locations. We can leverage these insights to tailor marketing strategies and meal offerings to match local preferences, thereby enhancing customer satisfaction and increasing order frequency.

### Visualization

```python
# Popular Dishes
plt.figure(figsize=(8, 6))
top_dishes.plot(kind='bar', color='skyblue')
plt.title('Top 5 Dishes by Order Count')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Top_Dishes.png")
plt.show()
```

```python
# Orders by Meal Type
plt.figure(figsize=(8, 6))
meal_type_counts = completed_sessions['Meal Type'].value_counts()
meal_type_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title('Orders by Meal Type')
plt.ylabel('')
plt.tight_layout()
plt.savefig("Orders_by_Meal_Type.png")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(rating_correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Session and Order Ratings')
plt.savefig("Rating_Correlation.png")
plt.show()
```
